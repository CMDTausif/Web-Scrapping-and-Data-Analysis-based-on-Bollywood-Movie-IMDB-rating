```python
# top 10 movies according to genre
top10 = data.sort_values(by="Genre", ascending=False).head(10)
top10
```

```python
# Top 10 movies according to year of release and genre
top10 = data.sort_values(by=["Year of release", "Genre"], ascending=False).head(10)
top10
```

```python
# Top movies as per the IMDB rating

data[data['Votes']>str(25000)].sort_values(by='Movie Rating', ascending=True)
```

```python
data.describe(include= 'all')
```

```python
data["Top_bollywood"] = data["Movie Rating"].rank(method="first", ascending= False)
data
```

```python
new_data = data.rename(columns={'Movie Rating': 'Movie_rate'})
new_data
```

```python
# top 10 directors and their score

new_data.groupby('Director').Movie_rate.mean().sort_values(ascending=True).head(10)
```

```python
# score of movie according to movie rating
new_data.groupby('Name of movie').Movie_rate.mean().sort_values(ascending=True).head(10)
```

```python
Top_10_popular_actor = new_data.sort_values(by='Name of movie', ascending=True, ignore_index=True).loc[0:10, ['Cast']].values.tolist()
Top_10_popular_actor
```

```python
Top_10_popular_movie = new_data.sort_values(by='Cast', ascending=True, ignore_index=True).loc[0:10, ['Name of movie']].values.tolist()
Top_10_popular_movie 
```
